# Create CSV file from multiple simulation experiments

```python
sim_root = "/Users/ashkanhzdr/workspace/infectio-mesa/output/trina_data/good_wo_superinf(dataC)"
output_file = "/Users/ashkanhzdr/workspace/infectio-mesa/output/trina_data/to_send/dataC.csv"
```

```python
import utils

paths = utils.list_sim_paths(sim_root)
paths
```

```python
import os
import pandas as pd
import numpy as np

# Initialize a dictionary to hold the data
consolidated_data = {
    't': None,  # Placeholder for 't' values
    'inf-count-list': [],
    'area-list': [],
    'radial-velocity-list': []
}

for path in paths:
    # Construct the file path to metric.csv
    metric_path = os.path.join(path, 'metric.csv')
    
    df = pd.read_csv(metric_path)
    
    # If 't' hasn't been set yet, initialize it from the first file
    if consolidated_data['t'] is None:
        consolidated_data['t'] = [round(v, 2) for v in df['t'].tolist()]
    else:
        # Ensure consistency of 't' across all files
        if not consolidated_data['t'] == [round(v, 2) for v in df['t'].tolist()]:
            raise ValueError(f"'t' values in {path} do not match previous files.")
    
    # Append values as lists to the respective keys in the dictionary
    consolidated_data['inf-count-list'].append(df['infected-count'].tolist())
    consolidated_data['area-list'].append(df['area(um2)'].tolist())
    consolidated_data['radial-velocity-list'].append(df['radial-velocity(um/min)'].tolist())

inf_count_transposed = list(map(list, zip(*consolidated_data['inf-count-list'])))
area_transposed = list(map(list, zip(*consolidated_data['area-list'])))
radial_velocity_transposed = list(map(list, zip(*consolidated_data['radial-velocity-list'])))

# Create a new DataFrame from the consolidated data
new_df = pd.DataFrame({
    't': consolidated_data['t'],
    'inf-count-list': inf_count_transposed,
    'area-list': area_transposed,
    'radial-velocity-list': radial_velocity_transposed
})

# Save the consolidated DataFrame to a new CSV file
new_df.to_csv(output_file, index=False)

print(f"Consolidated CSV file saved to {output_file}")


```
